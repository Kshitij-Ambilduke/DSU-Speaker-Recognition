```python
import pandas as pd

# Read the aligned files and create a mapping dictionary
tsv_file = "/mnt/data-artemis/kshitij/TEMP/Speaker-recognition/DSU_creation/VCTK/tsv_dir/audio_dev.tsv"
kmeans_file = "/mnt/data-artemis/kshitij/TEMP/Speaker-recognition/DSU_creation/VCTK/kmeans_dir_L22_C2000_output/audio_dev_0_1.km"

with open(tsv_file, 'r') as f1:
    f1_data = f1.readlines()
f1_data = f1_data[1:]
f1_data = [i.split('\t')[0] for i in f1_data]

with open(kmeans_file, 'r') as f2:
    f2_data = f2.readlines()
f2_data = [i.strip() for i in f2_data]
assert len(f1_data)==len(f2_data)

file_pairs = dict(zip(f1_data, f2_data))

print(len(file_pairs))
# Read your dataframe (replace this with your actual data loading method)
df = pd.read_pickle('/mnt/data-artemis/kshitij/TEMP/Speaker-recognition/useful_df/vctk/30_class_dev.pkl')  # Adjust parameters to match your file format

# Extract base filename from path and add new column
df['DSU'] = df['file'].apply(
    lambda x: file_pairs.get(x.split('/')[-1], None)
)

# Display the result
df
```
